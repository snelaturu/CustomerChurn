```python
import pandas as pd
import scipy.stats

telecom_df=pd.read_csv('Telco_Customer_Churn_Data.csv')

telecom_df
```

```python
telecom_df.dtypes
telecom_df.isnull().sum()
```

```python
# creating a data frame for checking the percentage of missing values

missing_value_percentage_df=pd.DataFrame({'Column_Name':telecom_df.columns,
                                          'Missing_value_count':telecom_df.isnull().sum(),
                                          'Missing_value_percentage':telecom_df.isnull().sum()/telecom_df.shape[0]*100})

missing_value_percentage_df=missing_value_percentage_df.sort_values(by='Missing_value_percentage',ascending=False)

missing_value_percentage_df.columns
```

```python
# dropping all the columns with missing vlaues > 50 % and 
# columns with unique values ever record and columns with same value with every record
cols_to_drop=[]

cols_with_na_grt_50=list(missing_value_percentage_df[missing_value_percentage_df['Missing_value_percentage']>50].index)

for col in telecom_df.columns:
    if telecom_df[col].nunique()==telecom_df.shape[0] or telecom_df[col].nunique()==1:
        cols_to_drop.append(col)
        
columns_to_drop=cols_to_drop+cols_with_na_grt_50

telecom_df.drop(columns=columns_to_drop,inplace=True)
```

```python
telecom_df['TotalCharges']=pd.to_numeric(telecom_df['TotalCharges'],errors='coerce')
#telecom_df.dtypes
telecom_df['SeniorCitizen']=telecom_df['SeniorCitizen'].astype(object)
```

```python
telecom_df.isnull().sum()
```

```python
telecom_df['TotalCharges'].fillna(0,inplace=True)
```

```python
cat_cols =  [key for key in dict(telecom_df.dtypes) if dict(telecom_df.dtypes)[key] in ['object']]
cont_cols = [key for key in dict(telecom_df.dtypes) if dict(telecom_df.dtypes)[key] not in ['object']]
```

```python
cat_cols
```

```python
cont_cols
```

```python
telecom_df['Churn'].replace('Yes',1,inplace=True)
```

```python
telecom_df['Churn'].replace('No',0,inplace=True)
```

```python
telecom_df
```

```python
# To identify the corelation between catagorical variables and continous variable
#ANOVA
import scipy.stats as stats
def anova(col):
    #categories=telecom_df[col].unique()
    return stats.f_oneway(telecom_df[col],telecom_df['Churn'])[1]
```

```python
anova_df=pd.DataFrame({'Column_Name':cont_cols,'p_value':[anova(col) for col in cont_cols ]})
anova_df
#The below value is zero that means it is extremly corelated
```

```python
# removing the output variable from the dataframe

y=telecom_df['Churn']

telecom_df.drop(columns='Churn',inplace=True)
```

```python
cat_cols.remove('Churn')
```

```python
#The chi-square test tests the null hypothesis that compares the correlation between categorical variables
pvalues_list=[]
for col in cat_cols:
    contengency_table = pd.crosstab(telecom_df[col],y)
    result=scipy.stats.chi2_contingency(contengency_table)
    pvalues_list.append(result[1])
```

```python
p_values_df=pd.DataFrame({'Column_name':cat_cols,'p_value':pvalues_list})
```

```python
p_values_df
# values <0.05 is correlated and > 0.05 is not correlated(eg:Gender,phoneService), if we want we can drop the not correlated columns
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(telecom_df,y,test_size=0.2,random_state=43)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#minmax scaling for continuous columns
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

minmaxscale = MinMaxScaler()
#stdscale = StandardScaler()
for col in cont_cols:
    X_train[col]= minmaxscale.fit_transform(np.array(X_train[col]).reshape(-1,1))
    X_test[col] = minmaxscale.transform(np.array(X_test[col]).reshape(-1,1))
    
# below are the warnings, you can ignore
```

```python
one_hot_train=pd.get_dummies(X_train[cat_cols]) # one-hot encoding
one_hot_test=pd.get_dummies(X_test[cat_cols]) # one-hot test
```

```python
#ensuring that noof category columns are same using inner join
X_train_oe,X_test_oe=one_hot_train.align(one_hot_test,join='inner',axis=1,fill_value=0)
```

```python
#merging the one hot encoded train data with the continous data
X_train_final=pd.concat([X_train[cont_cols],X_train_oe],axis=1)
X_train_final
```

```python
#merging the one hot encoded test data with the continous data
X_test_final=pd.concat([X_test[cont_cols],X_test_oe],axis=1)
X_test_final
```

```python
# Training the logistic regression model due to clasification probelm
from sklearn.linear_model import LogisticRegression

logreg=LogisticRegression()
logreg.fit(X_train_final,y_train)

```

```python
# Predections on the test data
logreg_pred=logreg.predict(X_test_final)
```

```python
# evaluate the model performance
from sklearn.metrics import confusion_matrix,f1_score,recall_score,precision_score

confusion_matrix(y_test,logreg_pred)
```

```python
y_test
```

```python
# below we got the results of 60% which is not a great model, need to improve the model
print('Precision-Score',precision_score(y_test,logreg_pred,pos_label='1'))
print('Recall-Score',recall_score(y_test,logreg_pred,pos_label='1'))
print('F1-Score',f1_score(y_test,logreg_pred,pos_label='1'))
```
